```python
# Importing packages
import numpy as np
import matplotlib.pyplot as plt

from fastchange.seg.pelt import PeltSeg
from fastchange.costs.normal import NormalMeanVarCost
from fastchange.penalties import mbic_penalty
```

```python
# Creating a synthetic dataset
x = np.hstack([
    np.random.normal(0, 1, (100,)),
    np.random.normal(10, 2, (100,)),
    np.random.normal(2, 2, (100,)),
    np.random.normal(-4, 1, (100,))
])

# Plotting artificial series
plt.figure()
plt.plot(x)
plt.show()
```

```python
# Running AMOC segmentation
amoc = PeltSeg(NormalMeanVarCost(), mbic_penalty).fit(x)
cps = amoc.predict()

# Plotting discovered changepoint
plt.figure()
plt.plot(x)
for cp in cps:
    plt.axvline(cp, color='red')
plt.show()
```
